```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

data = np.array([
    [15, 25, 125],
    [17, 29, 110],
    [12, 22, 105],
    [20, 30, 130],
    [18, 28, 120]
])

scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

pca = PCA(n_components=2)
data_baru = pca.fit_transform(data_normalized)

print("Data baru (setelah PCA):")
print(data_baru)
print(f"\nVariance ratio: {pca.explained_variance_ratio_}Z")
```
